```python
#loading data
import numpy as np
import scipy.io
GER_tr = scipy.io.loadmat('/content/drive/MyDrive/postprocessing/GER_train.mat');
GER_tt = scipy.io.loadmat('/content/drive/MyDrive/postprocessing/GER_test.mat');

ITA_tr = scipy.io.loadmat('/content/drive/MyDrive/postprocessing/ITA_train.mat');
ITA_tt = scipy.io.loadmat('/content/drive/MyDrive/postprocessing/ITA_test.mat');

GER_cr = scipy.io.loadmat('/content/drive/MyDrive/Context_features/Ger_context_train.mat');
GER_ct = scipy.io.loadmat('/content/drive/MyDrive/Context_features/Ger_context_test.mat');

ITA_cr = scipy.io.loadmat('/content/drive/MyDrive/Context_features/Ita_context_train.mat');
ITA_ct = scipy.io.loadmat('/content/drive/MyDrive/Context_features/Ita_context_test.mat');



GER_train = GER_tr['GER_train']
#GER_train=np.concatenate((GER_train,GER_cr["Ger_context_tr"][:-1]), axis=0)
GER_ytrain = GER_train[0]
GER_wtrain =GER_train[1]
GER_xtrain = GER_train[2:].T

GER_test = GER_tt['GER_test']
#GER_test=np.concatenate((GER_test,GER_ct["Ger_context_te"][:-1]), axis=0)
GER_ytest = GER_test[0]
GER_wtest = GER_test[1]
GER_xtest = GER_test[2:].T

ITA_train = ITA_tr['ITA_train']
#ITA_train=np.concatenate((ITA_train,ITA_cr["Ita_context_tr"][:-1]), axis=0)
ITA_ytrain = ITA_train[0]
ITA_wtrain = ITA_train[1]
ITA_xtrain = ITA_train[2:].T

ITA_test = ITA_tt["ITA_test"]
#ITA_test=np.concatenate((ITA_test,ITA_ct["Ita_context_te"][:-1]), axis=0)
ITA_ytest = ITA_test[0]
ITA_wtest = ITA_test[1]
ITA_xtest = ITA_test[2:].T

ita_size=ITA_train.shape
ger_size=GER_train.shape
ip1=ITA_train[:, :ita_size[1]//5]
ip2=ITA_train[:, (1*ita_size[1])//5:(2*ita_size[1])//5]
ip3=ITA_train[:, (2*ita_size[1])//5:(3*ita_size[1])//5]
ip4=ITA_train[:, (3*ita_size[1])//5:(4*ita_size[1])//5]
ip5=ITA_train[:, (4*ita_size[1])//5:]

gp1=GER_train[:, :ger_size[1]//5]
gp2=GER_train[:, (1*ger_size[1])//5:(2*ger_size[1])//5]
gp3=GER_train[:, (2*ger_size[1])//5:(3*ger_size[1])//5]
gp4=GER_train[:, (3*ger_size[1])//5:(4*ger_size[1])//5]
gp5=GER_train[:, (4*ger_size[1])//5:]

MIX_train=np.concatenate((ip1, gp1, ip2, gp2, ip3, gp3, ip4, gp4, ip5, gp5), axis=1)

MIX_ytrain = MIX_train[0]
MIX_wtrain = MIX_train[1]
MIX_xtrain = MIX_train[2:].T

print(GER_wtrain.shape)
print(ITA_wtrain.shape)
print(MIX_wtrain.shape)
print(MIX_ytrain.shape)
print(MIX_xtrain.shape)


```

```python
import numpy as np
#mixed
# Reshaping data
test_features_main = np.concatenate((ITA_xtest, GER_xtest), axis=0)
test_labels = np.concatenate((ITA_ytest, GER_ytest), axis=0)
test_words=np.concatenate((ITA_wtest, GER_wtest), axis=0)

train_features = MIX_xtrain
train_labels = MIX_ytrain
train_words = MIX_wtrain


train_size=train_features.shape[0]
language="mixed"
```

```python
test_features1_main = GER_xtest
test_labels1 = GER_ytest
test_words1 = GER_wtest

test_features2_main = ITA_xtest
test_labels2 = ITA_ytest
test_words2 = ITA_wtest

val_features5=train_features[(train_size*4)//5:]
train_features5=train_features[:((train_size*4)//5)]
val_labels5=train_labels[(train_size*4)//5:]
train_labels5=train_labels[:((train_size*4)//5)]
val_words5=train_words[:((train_size*4)//5)]
fold= "part 5"

val_features4=train_features[(train_size*3)//5:(train_size*4)//5]
train_features4=np.concatenate((train_features[:(train_size*3)//5], train_features[((train_size*4)//5):]), axis=0)
val_labels4=train_labels[(train_size*3)//5:(train_size*4)//5]
train_labels4=np.concatenate((train_labels[:(train_size*3)//5], train_labels[((train_size*4)//5):]), axis=0)
val_words4=train_words[(train_size*3)//5:(train_size*4)//5]
fold="part 4"

val_features3=train_features[(train_size*2)//5:(train_size*3)//5]
train_features3=np.concatenate((train_features[:(train_size*2)//5], train_features[((train_size*3)//5):]), axis=0)
val_labels3=train_labels[(train_size*2)//5:(train_size*3)//5]
train_labels3=np.concatenate((train_labels[:(train_size*2)//5], train_labels[((train_size*3)//5):]), axis=0)
val_words3=train_words[(train_size*2)//5:(train_size*3)//5]
fold="part 3"

val_features2=train_features[(train_size*1)//5:(train_size*2)//5]
train_features2=np.concatenate((train_features[:(train_size*1)//5], train_features[((train_size*2)//5):]), axis=0)
val_labels2=train_labels[(train_size*1)//5:(train_size*2)//5]
train_labels2=np.concatenate((train_labels[:(train_size*1)//5], train_labels[((train_size*2)//5):]), axis=0)
val_words2=train_words[(train_size*1)//5:(train_size*2)//5]
fold="part 2"

val_features1=train_features[:train_size//5]
train_features1=train_features[(train_size//5):]
val_labels1=train_labels[:train_size//5]
train_labels1=train_labels[(train_size//5):]
val_words1=train_words[:train_size//5]
fold="part 1"

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from keras.layers import Dense
from keras.callbacks import ModelCheckpoint ,EarlyStopping
from keras.layers import Dropout

from keras import backend as K
import keras
from sklearn import svm

def convert_function(arr):
  temp=[]
  for i in arr:
    if i>0.5:
      temp.append(1)
    else:
      temp.append(0)
  return temp

def calculate_accuracy(arr1, arr2):
  count=0
  for itr1, itr2 in zip(arr1, arr2):
    if itr1==itr2:
      count+=1
  return count/len(arr1)

def make_partitions(arr_words, arr_labels):
  v=[]
  np.array(v)
  temp=[]
  for i in range(len(arr_words)-1):
    word=arr_words[i]
    next_word=arr_words[i+1]
    temp.append(arr_labels[i][0])
    if word!=next_word:
      numpy_temp=np.array(temp)
      temp_max=np.amax(numpy_temp)
      numpy_temp=np.divide(numpy_temp, temp_max)
      v=np.concatenate((v, numpy_temp), axis=None)
      temp.clear()
    if (i==len(arr_words)-2):
      temp.append(arr_labels[i+1][0])
      numpy_temp=np.array(temp)
      temp_max=np.amax(numpy_temp)
      numpy_temp=np.divide(numpy_temp, temp_max)
      v=np.concatenate((v, numpy_temp), axis=None)
      temp.clear()
  v1=[]
  for i in v:
    if i==1:
      v1.append(1)
    else:
      v1.append(0)
  return v1

def sampling(args):
    z_mean, z_log_sigma = args
    epsilon = K.random_normal(shape=(K.shape(z_mean)[0], latent_dim), mean=0., stddev=0.004)
    return z_mean + K.exp(z_log_sigma) * epsilon

table=[]
table.append(["MIX_GER SVM before", "MIX_GER SVM after", "MIX_GER DNN before", "MIX_GER DNN after", "MIX_ITA SVM before", "MIX_ITA SVM after", "MIX_ITA DNN before", "MIX_ITA DNN after",])
for fold in ["part 1", "part 2", "part 3", "part 4", "part 5"]:
  if fold=="part 5":
    val_features=np.copy(val_features5)
    val_labels=val_labels5
    val_words=val_words5
    train_features=np.copy(train_features5)
    train_labels=train_labels5
  elif fold=="part 4":
    val_features=np.copy(val_features4)
    val_labels=val_labels4
    val_words=val_words4
    train_features=np.copy(train_features4)
    train_labels=train_labels4
  elif fold=="part 3":
    val_features=np.copy(val_features3)
    val_labels=val_labels3
    val_words=val_words3
    train_features=np.copy(train_features3)
    train_labels=train_labels3
  elif fold=="part 2":
    val_features=np.copy(val_features2)
    val_labels=val_labels2
    val_words=val_words2
    train_features=np.copy(train_features2)
    train_labels=train_labels2
  elif fold=="part 1":
    val_features=np.copy(val_features1)
    val_labels=val_labels1
    val_words=val_words1
    train_features=np.copy(train_features1)
    train_labels=train_labels1
  

  
  test_features=np.copy(test_features_main)
  test_features1=np.copy(test_features1_main)
  test_features2=np.copy(test_features2_main)


  avg_trainfeat=np.mean(train_features, axis=0)
  std_trainfeat=np.std(train_features, axis=0)
  i=0
  for v in train_features:
    train_features[i]=np.divide((v-avg_trainfeat), std_trainfeat)
    i=i+1
  i=0
  for v in val_features:
    val_features[i]=np.divide((v-avg_trainfeat), std_trainfeat)
    i=i+1
  i=0
  for v in test_features:
    test_features[i]=np.divide((v-avg_trainfeat), std_trainfeat)
    i=i+1
  i=0
  for v in test_features1:
    test_features1[i]=np.divide((v-avg_trainfeat), std_trainfeat)
    i=i+1
  i=0
  for v in test_features2:
    test_features2[i]=np.divide((v-avg_trainfeat), std_trainfeat)
    i=i+1
  i=0



  original_dim = 19
  intermediate_dim = 64
  latent_dim = 19

  lab_input = keras.Input(shape=(1,), name="lab_input")
  ae_input = keras.Input(shape=(19,), name="input")

  h = layers.Dense(64, activation='relu')(ae_input)
  z_mean = layers.Dense(19)(h)
  z_log_sigma = layers.Dense(19)(h)

  z = layers.Lambda(sampling)([z_mean, z_log_sigma])

#print (z.shape)
  encoder = keras.Model(inputs=[ae_input, lab_input], 
                        outputs=z
                        )
# Create decoder
  x = Dense(19, activation='relu')(z)
  x = Dense(64, activation='relu')(x)
  d_pred = Dense(19, activation='sigmoid', name="same")(x)

# instantiate VAE model
#outputs = decoder(encoder(inputs)[2])
#vae = keras.Model(inputs, outputs, name='vae_mlp')

  clf_features=Dense(128, activation='relu')(z)
  clf_features=Dropout(0.45)(clf_features)
  clf_features=Dense(64, activation='relu')(clf_features)
  clf_features=Dense(32, activation='relu')(clf_features)
  clf_features=Dense(16, activation='relu')(clf_features)
  clf_features=Dense(8, activation='relu')(clf_features)
  clf_features=Dense(4, activation='relu')(clf_features)
  clf_features=Dense(2, activation='relu')(clf_features)
  clf_pred=Dense(1, activation='sigmoid', name='stress')(clf_features)

  reconstruction_loss = keras.losses.mean_squared_error(ae_input, d_pred)
  reconstruction_loss *= 19
  kl_loss = 1 + z_log_sigma - K.square(z_mean) - K.exp(z_log_sigma)
  kl_loss = K.sum(kl_loss, axis=-1)
  kl_loss *= -0.5
  vae_loss = K.mean(reconstruction_loss + kl_loss)

  l1=keras.losses.binary_crossentropy(lab_input ,clf_pred)
  clf_loss = K.mean(l1)

  model = keras.Model(
      inputs=[ae_input, lab_input],
      outputs=[d_pred, clf_pred],
  )
  model.add_loss({"same":vae_loss})
  model.add_loss({"stress":clf_loss})
  model.compile(
      optimizer='adam',
      loss_weights={"same": 0.3, "stress":1},
  )

  callbacks=[EarlyStopping(monitor='loss', mode='min', verbose=1, patience=10)]

  model.fit({"input": train_features, "lab_input":train_labels}, {"same": train_features, "stress": train_labels}
            , epochs=500, batch_size=128
            , validation_data=({"input": val_features, "lab_input": val_labels}, {"same": val_features, "stress": val_labels})
            , callbacks=callbacks
            )

  string="/content/drive/MyDrive/models/vaecdnn"+language+fold+".h5"
  model.save(string) 
  string="/content/drive/MyDrive/models/vaecdnn+encoder"+language+fold+".h5"
  encoder.save(string) 


  pred_output=model([test_features, test_labels])
  print(pred_output[1])
  pred_labels=pred_output[1].numpy()
  print(len(test_labels))

  pred1_labels=pred_output[1].numpy()
  for i in range (len(pred1_labels)):
    if pred1_labels[i]>0.5:
      pred1_labels[i]=1
    else:
      pred1_labels[i]=0
  print(pred1_labels)


  post_labels=make_partitions(test_words, pred_labels)
  print(calculate_accuracy(pred1_labels, test_labels))
  print(calculate_accuracy(post_labels, test_labels))
  svm_train=encoder([train_features, train_labels])
  svm_test=encoder([test_features, test_labels])

  clf= svm.SVC(probability=True)
  clf.fit(svm_train, train_labels)
  pred_svm_labels= clf.predict(svm_test)
  prob_labels=clf.predict_proba(svm_test)
  prob_labels=np.hsplit(prob_labels, 2)[1]
  post_svm_labels=make_partitions(test_words, prob_labels)
  print(calculate_accuracy(pred_svm_labels, test_labels))
  print(calculate_accuracy(post_svm_labels, test_labels))
  print(language, fold)
  print("vae+cdnn")
  ger_svm_test=encoder([test_features1, test_labels1])
  pred_svm_ger_labels=clf.predict(ger_svm_test)
  prob_ger_labels=clf.predict_proba(ger_svm_test)
  prob_ger_labels=np.hsplit(prob_ger_labels, 2)[1]
  post_svm_ger_labels=make_partitions(test_words1, prob_ger_labels)
  var1=calculate_accuracy(pred_svm_ger_labels, test_labels1)
  var2=calculate_accuracy(post_svm_ger_labels, test_labels1)
  print(var1)
  print(var2)

  pred_ger_output=model([test_features1, test_labels1])
  pred_ger_labels=pred_ger_output[1].numpy()
  pred1_ger_labels=pred_ger_output[1].numpy()
  pred1_ger_labels=convert_function(pred1_ger_labels)
  post_ger_labels=make_partitions(test_words1, pred_ger_labels)
  var3=calculate_accuracy(pred1_ger_labels, test_labels1)
  var4=calculate_accuracy(post_ger_labels, test_labels1)
  print(var3)
  print(var4)
  print("german test svm, clf", language, fold)


  ita_svm_test=encoder([test_features2, test_labels2])
  pred_svm_ita_labels=clf.predict(ita_svm_test)
  prob_ita_labels=clf.predict_proba(ita_svm_test)
  prob_ita_labels=np.hsplit(prob_ita_labels, 2)[1]
  post_svm_ita_labels=make_partitions(test_words2, prob_ita_labels)
  var5=calculate_accuracy(pred_svm_ita_labels, test_labels2)
  var6=calculate_accuracy(post_svm_ita_labels, test_labels2)
  print(var5)
  print(var6)
  pred_ita_output=model([test_features2, test_labels2])
  pred_ita_labels=pred_ita_output[1].numpy()
  pred1_ita_labels=pred_ita_output[1].numpy()
  pred1_ita_labels=convert_function(pred1_ita_labels)
  post_ita_labels=make_partitions(test_words2, pred_ita_labels)
  var7=calculate_accuracy(pred1_ita_labels, test_labels2)
  var8=calculate_accuracy(post_ita_labels, test_labels2)
  print(var7)
  print(var8)
  table.append([var1, var2, var3, var4, var5, var6, var7, var8])
print(language)
```
